```python
#All unique values of each columns
print("Ship Mode=",df['Ship Mode'].unique())
print("Country=",df['Country'].unique())
print("Category",df['Category'].unique())
print("State",df['State'].unique())
print("Segment",df['Segment'].unique())
print("Sub-Category=",df['Sub-Category'].unique())
```

```python
df.groupby('Region')['Sales'].sum().plot.bar()
plt.ylabel('Sales')
plt.title('Total Sales by Region')
plt.show()
```

```python
df.groupby('Region')['Profit'].sum().plot.bar()
plt.ylabel('Profit')
plt.title('Total Profit by Region')
plt.show()
```

```python
df.groupby("Category")['Sales'].sum().plot.bar(color=["orange","purple","brown"])
plt.title("Category Vs Sales")
plt.ylabel('Sales')
plt.show()
```

```python
df.groupby("Category")['Profit'].sum().plot.bar(color=["orange","purple","brown"])
plt.title("Category Vs Profit")
plt.ylabel("Profit")
plt.show()
```

```python
df.groupby("Segment")['Sales'].sum().plot.bar(color=['green', 'blue', 'red'])
plt.title("Segment Vs Sales")
plt.ylabel('Sales')
plt.show()
```

```python
df.groupby("Segment")['Profit'].sum().plot.bar(color=['green', 'blue', 'red'])
plt.title("Segment Vs Profit")
plt.ylabel('Profit')
plt.show()
```

```python
df.groupby('State')['Sales'].sum().plot.bar()
plt.title('States vs Sales')
plt.ylabel('Sales')
plt.show()
```

```python
df.groupby('State')['Profit'].sum().plot.bar()
plt.title('States vs Profit')
plt.ylabel('Profit')
plt.show()
```
